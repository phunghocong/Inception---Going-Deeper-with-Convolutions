```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.utils import plot_model
```

```python
from Inception import InceptionFramework
Inception = InceptionFramework()
```

```python
# define model input
in_layer = Input(shape=(256, 256, 3))
# add inception module
incep_block = Inception.Naive_Inception_Module(in_layer, 64, 128, 32)
# create model
model = Model(inputs = in_layer, outputs = incep_block)
# summarize model
model.summary()
```

```python
plot_model(model, show_shapes=True, to_file='vanilla_inception.png')
```
